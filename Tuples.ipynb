### Tuples 

- similar to list but this is immutable


```python
tup = (23,45,22,'b.tech')
tup
```

```python
type(tup)
```

```python
tup[2] # accessing the values
```

- to create a empty tuple

```python
tup1 = ()
tup1
```

```python
# if you have only 1 val in tup then 
tup2 = (23,)
tup2
```

```python
type(tup2)
```

```python
# because if you don't use , then interpreter will take it as normal data-type
tup3 = (23)
type(tup3)
```

- changing the value gives you error

```python
tup[2] = 34
tup
```

- same concept of slicing exist in tuple too

```python
tup = (1,2,3,4,5)
tup[2:4]
```

### methods in tuples

- index()
- count()

```python
tup = (1,2,3,4,1,2,5,2)

tup.index(1)
```

```python
tup.count(2)
```
