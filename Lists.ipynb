### List Slicing

- list_name[start_idx : ending_idx : step]
- if you don't specify ending_idx it goes to len(list)
- ending_idx is not included
- step if not specified it is normally 1

```python
marks = [99,34,56,78,85,21,34,56]
marks[1:4]
```

```python
marks[1::2]
```

- negative indexing in list

```python
marks[-5:-1]
```

```python
marks[-1:]
```

```python
marks[:-1]
```

```python
marks[::-1] # reversing the list
```
